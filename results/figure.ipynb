```python
import matplotlib.pyplot as plt
import numpy as np

# Data
categories = ['base task', 'base task\n+ 1 sub-task', 'base task\n+ 2 sub-tasks',
              'base task\n+ 3 sub-tasks', 'base task\n+ 4 sub-tasks', 'base task\n+ 5 sub-tasks']
baseline = [20, 22, 26, 42, 51, 56]
task_fusion = [21, 24, 23, 47, 53, 60]
task_fusion_scheduling = [18, 34, 43, 53, 64, 67]
spatten_task_fusion = [19, 41, 50, 55, 60, 60]
spatten_task_fusion_scheduling = [20, 36, 50, 64, 67, 67]

# Percentages
percentages = ['26.8%', '22.0%', '36.8%', '42.6%', '51.9%',
               '28.2%', '24.1%', '44.2%', '53.0%', '56.0%',
               '33.1%', '36.1%', '52.6%', '64.1%', '60.3%',
               '36.1%', '50.0%', '55.6%', '67.7%']

# Create figure and axis
fig, ax = plt.subplots(figsize=(10, 5))

# Width of a bar 
width = 0.15       

# Position of bars on x-axis
ind = np.arange(len(baseline))

# Plotting
ax.bar(ind, baseline, width, label='Baseline', color='tan')
ax.bar(ind + width, task_fusion, width, label='TaskFusion', color='lightcoral')
ax.bar(ind + 2*width, task_fusion_scheduling, width, label='TaskFusion+scheduling', color='lightblue')
ax.bar(ind + 3*width, spatten_task_fusion, width, label='SpAtten+TaskFusion', color='lightgreen')
ax.bar(ind + 4*width, spatten_task_fusion_scheduling, width, label='SpAtten+TaskFusion+scheduling', color='lightgrey')

# Adding percentages on top of the bars
for i, rect in enumerate(ax.patches):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width()/2., 1.05*height,
            percentages[i % len(categories)],
            ha='center', va='bottom', rotation='vertical', color='black')

# X-axis labels
ax.set_xticks(ind + 2*width)
ax.set_xticklabels(categories)

# Y-axis label
ax.set_ylabel('System Latency [ms]')

# Title
ax.set_title('(a) System Latency')

# Legend
ax.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import numpy as np
import pandas as pd
import torch

# Data
# 데이터 프레임을 생성합니다.
shape = [32, 1024]
tensor = torch.rand(shape, dtype=torch.float16)
data = tensor.numpy()

x_pos, y_pos = np.meshgrid(np.arange(shape[0]), np.arange(shape[1]))
x_pos = x_pos.flatten()
y_pos = y_pos.flatten()
z_pos = np.zeros(shape[0]*shape[1])

z_size = data.flatten()

dx = dy = 0.1
dz = z_size
colors = cm.viridis(z_size / float(max(z_size)))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.bar3d(x_pos, y_pos, z_pos, dx, dy, dz, color=colors)
ax.view_init(90, 0)

# 라벨과 타이틀 추가
ax.set_xlabel('token')
ax.set_ylabel('channel')
ax.set_zlabel('Z Axis')
ax.set_title('3D Bar Graph of 2D Tensor')

plt.show()
```
